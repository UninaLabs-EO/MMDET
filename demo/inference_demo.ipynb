```python
from mmdet.apis import init_detector, inference_detector, show_result_pyplot
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import mmcv
```

```python
config_file = '/home/sirbastiano/Documenti/Scripts/MMDETv2/mmdetection/checkpoints/hrnet402022-07-06T08:56:37_712_40e_B4_lr_0.01/HrNet_w40_cascade.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
# url: https://download.openmmlab.com/mmdetection/v2.0/faster_rcnn/faster_rcnn_r50_fpn_1x_coco/faster_rcnn_r50_fpn_1x_coco_20200130-047c8118.pth
checkpoint_file = '/home/sirbastiano/Documenti/Scripts/MMDETv2/mmdetection/checkpoints/hrnet402022-07-06T08:56:37_712_40e_B4_lr_0.01/epoch_40.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single image
images = Path('/home/sirbastiano/Documenti/Scripts/MMDETv2/mmdetection/data/coco/val2017').glob('**/*')
images = [x for x in images if x.is_file()]
idx = np.random.randint(0, len(images))
img = images[idx]
result = inference_detector(model, img)
# show the results
plt.figure(figsize=(20,20))
show_result_pyplot(model, img, result)
```
